```python
import json
```

```python
#load the data, data is inconsistent, missing data, duplicate data
def data_loading(filename):
    with open(filename,"r") as f:
        data = json.load(f)
        
    return data
```

```python
data = data_loading("store_data.json")
print(type(data), data)
```

```python
#assignment1
#clean &  structure data
def clean_data(data):
    text_to_num ={"one":1, "two":2, "three":3, "four": 4, "five":5}
    cleaned_data =[]
    unique_users = set()
    for user in data:
        #clean ratings
        raw_rating = str(user["rating"]).strip().lower()
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        user["rating"]=raw_rating

        #handle missing vals
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None
        #deduplication
        if(user["name"].strip() in unique_users):
            continue

        unique_users.add(user["name"])
        cleaned_data.append(user)
    return cleaned_data
```

```python
clean_data(data)
```

```python
#assignment2
#Get meaningful insights from data
def get_insights(data):

    #avg rating
    tot_rating = 0
    for user in data:
        tot_rating+= float(user["rating"])
    print(f"avg rating = {tot_rating/len(data)}")

    #percentage of users with poor rating
    poor_rating = 0
    for user in data:
        if(float(user["rating"])<3):
            poor_rating +=1
            
    print(f"percentage of poor rating ={(poor_rating/len(data))*100}")
        
```

```python
get_insights(data)
```

```python
#assignment3
#built recommendatiob feature
def get_recommendations(data):
    recommendations = []

    for user in data:
        curr_recom = {}
        curr_recom["name"] = user["name"]
        if(float(user["rating"])>=4):
            curr_recom["brand"] = "Apple"
        else:
            curr_recom["brand"] = "Samsung"
        recommendations.append(curr_recom)
    return recommendations
            
```

```python
get_recommendations(data)
```
